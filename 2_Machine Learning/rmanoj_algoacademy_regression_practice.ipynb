```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# step 2 : Reading the dataset

```python
df = pd.read_csv(filepath_or_buffer='/Users/manojravikumar/Downloads/algoacademy.ai/loan_prediction_dataset.csv')
df.head()
```

# Step 3 : Seperating the Features and Target

```python
X=df[["Age","Annual_Income","Credit_Score"]]
X

y=df["Loan_Amount"]
y
```

# Step 4 : Splitting the dataset into training set and test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y , random_state=104,test_size=0.25, shuffle=True)
```

# Step 5 : Creating the Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
```

# Step 6 : Training the Model

```python
model.fit(X_train,y_train)
```

# Step 7 : Testing the Model

```python
y_pred = model.predict(X_test)
y_pred
```

```python
df1 = X_test.copy()
df1["Actual_Loan_Amount"] = y_test
df1["Predicted_Loan_Amount"] = y_pred

df1.head()
```
