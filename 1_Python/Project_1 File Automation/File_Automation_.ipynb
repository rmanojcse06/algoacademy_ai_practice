## Collecting unique extension folders

```python
import os
def fileorganizer(path):
    extensions = []
    files = os.listdir(path)
    for i in files:
        x = os.path.join(mypath, os.path.splitext(i)[1][1:])
        if x not in extensions:
            extensions.append(x)
    print(extensions)

mypath = r"/home/siva/Desktop/File Automation"
fileorganizer(mypath)
```

## Creating folders safely using try / except

```python
import os
def fileorganizer(path):
    extensions = []
    files = os.listdir(path)
    for i in files:
        x = os.path.join(mypath, os.path.splitext(i)[1][1:])
        if x not in extensions:
            extensions.append(x)

    for folder in extensions:
        try:
            os.mkdir(folder)
        except:
            pass
        print(os.path.split(folder)[1])

mypath = r"/home/siva/Desktop/File Automation"
fileorganizer(mypath)
```

## Moving files into respective folders

```python
import os

def fileorganizer(path):
    extensions = []
    files = os.listdir(path)

    for i in files:
        x = os.path.join(mypath, os.path.splitext(i)[1][1:])
        if x not in extensions:
            extensions.append(x)

    for folder in extensions:  # to create folders
        try:
            os.mkdir(folder)
        except:
            pass

        for file in files:
            if os.path.split(folder)[1] == os.path.splitext(file)[1][1:]:
                # (C:\Users\Siva\Desktop\Automation , pdf)

                os.rename(
                    os.path.join(path, file),# C:\Users\Siva\Desktop\Automation + arun.py
                    os.path.join(folder, file) # C:\Users\Siva\Desktop\Automation\pdf + arun.py
                )
                
               


mypath = r"/home/siva/Desktop/File Automation (Copy 2)"
fileorganizer(mypath)
```

# Explanation how OS.rename moves the file 

```python
os.rename(os.path.join(path, file), os.path.join(folder, file))

🔎 Step 1: os.path.join(path, file)
os.path.join() safely builds a full file path (it adds the correct / or \ depending on your OS).

Example:

path = "C:/Users/Siva/Documents"
file = "report.pdf"
              #"C:/Users/Siva/Documents" + "report.pdf"
os.path.join(path, file)  
# 👉 "C:/Users/Siva/Documents/report.pdf"
So this gives the current location of the file.
                      
🔎 Step 2: os.path.join(folder, file)
Same thing, but with the new folder path.

Example:
folder = "C:/Users/Siva/Documents/PDFs"
file = "report.pdf"
            #"C:/Users/Siva/Documents/PDFs" + ""report.pdf"
os.path.join(folder, file)  
# 👉 "C:/Users/Siva/Documents/PDFs/report.pdf"


This gives the destination path where the file should go.

🔎 Step 3: os.rename(src, dst)

Now, os.rename() moves the file:

src = os.path.join(path, file) → original location #C:/Users/Siva/Documents/report.pdf"
dst = os.path.join(folder, file) → new location #"C:/Users/Siva/Documents/PDFs/report.pdf"

So the code means:
👉 Take file from path and move/rename it into folder with the same name.

✅ Example
import os

path = "C:/Users/Siva/Documents"
folder = "C:/Users/Siva/Documents/PDFs"
file = "report.pdf"

os.rename(os.path.join(path, file), os.path.join(folder, file))


This moves:

C:/Users/Siva/Documents/report.pdf
      ↓
C:/Users/Siva/Documents/PDFs/report.pdf
```
