```python
import os
from openai_key import OPENAI_API_KEY  # import your key from openai_key.py

# LangChain imports
from langchain_openai import OpenAI
from langchain_core.prompts import PromptTemplate
from langchain_classic.chains import LLMChain

# Set the API key for LangChain/OpenAI
os.environ["OPENAI_API_KEY"] = OPENAI_API_KEY


def generate_startup_name(industry: str) -> dict:
    llm = OpenAI(temperature=0.6)

    # ------------------------------------------------------------
    # 🧩 Prompt Template for Startup Name
    # ------------------------------------------------------------
    startup_name_prompt = PromptTemplate(
        input_variables=["industry"],
        template="Generate a unique, catchy, brandable startup name for a company in the {industry} industry."
    )

    # ------------------------------------------------------------
    # 🔗 LLMChain to generate the startup name
    # ------------------------------------------------------------
    startup_name_chain = LLMChain(
        llm=llm,
        prompt=startup_name_prompt,
        output_key="startup_name"
    )

    # Invoke chain
    result = startup_name_chain.invoke({"industry": industry})
    return result


# Example usage
if __name__ == "__main__":
    output = generate_startup_name("Health Care")
    print(output)
```
