```python
# =========================================
# SMOTE DEMO ON IMBALANCED LOAN DATASET
# =========================================

import pandas as pd
import numpy as np

# -------------------------------
# 1. Load the dataset
# -------------------------------
df = pd.read_csv("loan_imbalanced_dataset.csv")

#print("First 5 rows:")
#print(df.head())

#print("\nOriginal class distribution:")
print(df["loan_status"].value_counts())
```

```python
# -------------------------------
# 2. Separate features & target
# -------------------------------
X = df[["age", "salary", "cibil_score"]]
y = df["loan_status"]


# -------------------------------
# 3. Train–Test Split
# (VERY IMPORTANT – no leakage)
# -------------------------------
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    stratify=y,
    random_state=42
)

print("\nTraining class distribution (before SMOTE):")
print(y_train.value_counts())
```

```python
# -------------------------------
# 4. Apply SMOTE (TRAIN DATA ONLY)
# -------------------------------
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=42)

X_train_smote, y_train_smote = smote.fit_resample(
    X_train,
    y_train
)

print("\nTraining class distribution (after SMOTE):")
print(pd.Series(y_train_smote).value_counts())
```

```python
# -------------------------------
# 5. Train a Model
# -------------------------------
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(max_iter=1000)
model.fit(X_train_smote, y_train_smote)


# -------------------------------
# 6. Evaluate on Test Data
# -------------------------------
from sklearn.metrics import classification_report

y_pred = model.predict(X_test)

print("\nModel Performance on Test Data:")
print(classification_report(y_test, y_pred))


# -------------------------------
# 7. Key Takeaway
# -------------------------------
print("""
IMPORTANT NOTES:
- SMOTE is applied ONLY on training data
- Test data is NEVER oversampled
- Accuracy alone is misleading for imbalanced data
- Focus on Recall & Precision for minority class
""")
```
